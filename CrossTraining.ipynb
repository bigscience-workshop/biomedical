# Equate labels
Create lookup dictionary to translate BRONCO and Cardio:DE labels to the GGPONC2 ones.

```python
lookup = {
    # for bronco
    'DIAGNOSIS': 'Finding',
    'TREATMENT': 'Procedure',
    'MEDICATION': 'Medication',
    
    # for cardiode
    'ACTIVEING': 'Medication',
    'DRUG': 'Medication',
}
```

```python
for row in bronco['train']:
    for ent in row['entities']:
        ent['type'] = lookup[ent['type']]
    
for row in cardiode['train']:
    for ent in row['entities']:
        if ent['type'] in lookup:
            ent['type'] = lookup[ent['type']]
```
